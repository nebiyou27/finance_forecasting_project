```python
import matplotlib.pyplot as plt
import seaborn as sns

def analyze_distribution(df):
    """Analyzes the distribution of numerical columns."""
    numerical_cols = df.select_dtypes(include=np.number).columns
    for col in numerical_cols:
        print(f"\n--- {col} ---")
        print(df[col].describe())

        plt.figure(figsize=(12, 6))
        plt.subplot(1, 2, 1)
        sns.histplot(df[col], kde=True)
        plt.title(f"Histogram of {col}")

        plt.subplot(1, 2, 2)
        sns.boxplot(y=df[col])
        plt.title(f"Boxplot of {col}")
        plt.show()

analyze_distribution(df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data (if you haven't already)
data_path = "C:/Users/neba/Desktop/finance_forecasting_project/data/raw/combined_financial_data.csv"
df = pd.read_csv(data_path, parse_dates=['Date'])

def detect_outliers(df, threshold=3):
    """Detects and visualizes outliers using z-score."""
    numerical_cols = df.select_dtypes(include=np.number).columns
    outlier_counts = {}

    for col in numerical_cols:
        z_scores = np.abs((df[col] - df[col].mean()) / df[col].std())
        outliers = df[z_scores > threshold]
        outlier_counts[col] = len(outliers)

        print(f"\n--- Outliers in {col} (Threshold: {threshold} standard deviations) ---")
        print(f"Number of outliers: {len(outliers)}")

        plt.figure(figsize=(10, 5))
        sns.boxplot(x=df[col])
        plt.title(f"Boxplot of {col} with Outliers")
        plt.show()

        if not outliers.empty:
            print("Outlier values:")
            print(outliers[col])

    print("\n--- Summary of Outlier Counts ---")
    for col, count in outlier_counts.items():
        print(f"{col}: {count} outliers")

detect_outliers(df)
```

```python
def clean_column_names(df):
    """Cleans up the column names."""
    new_columns = []
    for col in df.columns:
        if col == 'Date':
            new_columns.append('Date')
        else:
            asset, metric_tuple = col.split('_', 1)
            metric = metric_tuple.replace("('", "").replace("', '", "_").replace("')", "")
            new_columns.append(f"{asset}_{metric}")
    df.columns = new_columns
    print("\n--- Cleaned Column Names ---")
    print(df.columns)
    return df

df = clean_column_names(df)
print(df.head())
```

```python
def clean_column_names(df):
    """Cleans up the column names."""
    new_columns = []
    for col in df.columns:
        if col == 'Date':
            new_columns.append('Date')
        else:
            asset, metric_tuple = col.split('_', 1)
            metric = metric_tuple.replace("('", "").replace("', '", "_").replace("')", "")
            metric = metric.replace(f"_{asset}", "") # remove redundant asset name
            new_columns.append(f"{asset}_{metric}")
    df.columns = new_columns
    print("\n--- Cleaned Column Names ---")
    print(df.columns)
    return df

df = clean_column_names(df)
print(df.head())
```

```python
from sklearn.preprocessing import MinMaxScaler

def scale_data(df, columns_to_scale=None):
    """Scales numerical data using MinMaxScaler."""
    if columns_to_scale is None:
        columns_to_scale = df.select_dtypes(include=np.number).columns.tolist()

    scaler = MinMaxScaler()
    df[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])
    print("\n--- Data Scaled using MinMaxScaler ---")
    return df

# Select the numerical columns to scale (excluding 'Date')
numerical_cols = df.select_dtypes(include=np.number).columns.tolist()

df = scale_data(df, columns_to_scale=numerical_cols)

print(df.head())
```
